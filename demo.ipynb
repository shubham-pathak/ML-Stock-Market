```python
#n is 50
import matplotlib
import csv
import statistics as stat
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
with open("a.csv") as csvfile:
    readCSV= csv.reader(csvfile)
    vol=[]
    close_price=[]
    time=[]
    for i in readCSV:
        price=i[1]
        date= i[0]

        close_price.append(price)
        time.append(date)

close_price.remove(close_price[0]);
time.remove(time[0]);
        
X= np.array(time).astype('float')
y= np.array(close_price).astype('float')
plt.plot(y)
plt.title('Price Plot')
plt.ylabel('Price')
plt.xlabel('Period')
plt.show()
#print(y)
print(len(y))
```

```python
N=50
cumsum, moving_aves = [0], []
for i, x in enumerate(y, 1):
    cumsum.append(cumsum[i-1] + x)
    if i>=N:

        moving_ave = (cumsum[i] - cumsum[i-N])/N
#         print(cumsum)
#         print(moving_ave)

        #can do stuff with moving_ave here
        moving_aves.append(round(moving_ave,2))
# print(moving_aves)
l1=len(moving_aves)
#print(l2)
print(type(moving_aves))
sma=np.array(moving_aves).astype('float')
print(sma)
print(len(sma))
```

```python
diff=np.empty(493,float)
for i in range(0,493):
    diff[i]=y[i+1]-y[i]
print(diff)
print(len(diff))
```

```python
#RSI N=14
pos=np.zeros(493,float)
neg=np.zeros(493,float)
for i in range(0,493):
    if diff[i]>0 :
        pos[i]=diff[i]
    else:
        neg[i]=diff[i]
print(pos)
print(neg)
```

```python
rp=np.empty(494-13,float)
rn=np.empty(494-13,float)
rp[0]=stat.mean(pos[0:13])
rn[0]=stat.mean(neg[0:13])
```

```python
for i in range(1,494-13):
    rp[i]=((rp[i-1]*13)+pos[i])/14
    rn[i]=((rn[i-1]*13)+neg[i])/14
    print(i)
print(rp)
print(rn)


```

```python
rsi=np.empty(494-13,float)
for i in range(0,494-13):
    rs=rp[i]/rn[i]
    rsi[i]=100-(100/(1-rs))
print(len(rsi))
```
